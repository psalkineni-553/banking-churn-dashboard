```python
import pandas as pd

# Load the dataset
file_path = "../Data/Churn_Modelling.csv"
df = pd.read_csv(file_path)

# Basic inspection
print("Dataset shape:", df.shape)

print("\nColumn names:")
print(df.columns.tolist())

print("\nFirst 5 rows:")
display(df.head())

print("\nData types:")
display(df.dtypes)

print("\nMissing values:")
display(df.isnull().sum())

print("\nBasic summary:")
display(df.describe())
```

```python
# Show all column names with sample values
for column in df.columns:
    print(f"\nColumn: {column}")
    print(df[column].head(10).tolist())
```

```python
# Check churn distribution
churn_counts = df["Exited"].value_counts()
churn_rate = df["Exited"].mean() * 100

print("Churn counts:")
print(churn_counts)

print(f"\nOverall churn rate: {churn_rate:.2f}%")
```

```python
# Churn rate by geography
geo_churn = (
    df.groupby("Geography")["Exited"]
    .agg(["count", "sum", "mean"])
    .reset_index()
)

geo_churn["churn_rate_percent"] = geo_churn["mean"] * 100

print("Churn by Geography:")
display(geo_churn.sort_values("churn_rate_percent", ascending=False))
```

```python
# Churn rate by gender
gender_churn = (
    df.groupby("Gender")["Exited"]
    .agg(["count", "sum", "mean"])
    .reset_index()
)

gender_churn["churn_rate_percent"] = gender_churn["mean"] * 100

print("Churn by Gender:")
display(gender_churn.sort_values("churn_rate_percent", ascending=False))
```

```python
# Churn rate by active membership
active_churn = (
    df.groupby("IsActiveMember")["Exited"]
    .agg(["count", "sum", "mean"])
    .reset_index()
)

active_churn["churn_rate_percent"] = active_churn["mean"] * 100

print("Churn by Active Membership:")
display(active_churn.sort_values("churn_rate_percent", ascending=False))
```

```python
# Churn rate by number of products
products_churn = (
    df.groupby("NumOfProducts")["Exited"]
    .agg(["count", "sum", "mean"])
    .reset_index()
)

products_churn["churn_rate_percent"] = products_churn["mean"] * 100

print("Churn by Number of Products:")
display(products_churn.sort_values("NumOfProducts"))
```

```python
# Create age groups
df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=[0, 30, 40, 50, 60, 100],
    labels=["Under 30", "30-40", "41-50", "51-60", "60+"]
)

# Churn rate by age group
age_churn = (
    df.groupby("AgeGroup", observed=False)["Exited"]
    .agg(["count", "sum", "mean"])
    .reset_index()
)

age_churn["churn_rate_percent"] = age_churn["mean"] * 100

print("Churn by Age Group:")
display(age_churn)
```

```python
# Churn rate by credit card ownership
card_churn = (
    df.groupby("HasCrCard")["Exited"]
    .agg(["count", "sum", "mean"])
    .reset_index()
)

card_churn["churn_rate_percent"] = card_churn["mean"] * 100

print("Churn by Credit Card Ownership:")
display(card_churn.sort_values("churn_rate_percent", ascending=False))
```

```python
# Create balance groups
df["BalanceGroup"] = pd.cut(
    df["Balance"],
    bins=[-1, 0, 50000, 100000, 150000, 250000],
    labels=["Zero Balance", "1-50K", "50K-100K", "100K-150K", "150K+"]
)

# Churn rate by balance group
balance_churn = (
    df.groupby("BalanceGroup", observed=False)["Exited"]
    .agg(["count", "sum", "mean"])
    .reset_index()
)

balance_churn["churn_rate_percent"] = balance_churn["mean"] * 100

print("Churn by Balance Group:")
display(balance_churn)
```

```python
# Churn rate by tenure
tenure_churn = (
    df.groupby("Tenure")["Exited"]
    .agg(["count", "sum", "mean"])
    .reset_index()
)

tenure_churn["churn_rate_percent"] = tenure_churn["mean"] * 100

print("Churn by Tenure:")
display(tenure_churn.sort_values("Tenure"))
```
